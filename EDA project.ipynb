```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Business dataset.xlsx")
df.head()
```

```python
df.shape
```

From the dataset imported, we can see that the dataset has `3326` rows and `16` columns. Now to explore if the dataset needs additional cleaning or preprocessing.

The next thing we will proceed to do is to do is to drop the `Order ID` column due to it not having any major significance in generating major insights for our report.

```python
df = df.drop('Order ID',axis = 1)
```

```python
df.info()
```

There are no missing values across all the columns, hence there is no need for preprocessing or cleaning.

```python
df.describe()
```

```python
df.corr()
```

As we can see, the categories with the highest correlation is `Revenue` against `Sales quantity`.

This is a brief description of our dataset numerical values.

### Use Pandas Profiling for the EDA process

```python
from pandas_profiling import ProfileReport

profile = ProfileReport(df, title="EDA Report for Business dataset", html={'style':{'full-width':True}})
```

```python
profile
```

```python
profile.to_file("business_dataset_EDA.html")
```

From the above report, we can see that there were no missing values and the data was ready for analysis without any form of cleaning or preprocessing. The report above gives a basic report about the different columns we have to observe in our dataset. There were intuitive visualizations that were generated for each of these columns in order to explore what they were all about. 
- The most `Generic Name` was `Pregabalin` by a huge margin.  
- `Lyrica` was the `Brand Name` with the highest occurence. 
- `Pfizer` was also the prefered and dominant `Company Name` in our dataset.
- Most of the customers had `Neuropathic pain` under the `Disease Medical Use` category. 
- There was nothing abnormal about the `Invoice date`, as the distribution was average. 
- `India` and `Thailand` are the most preferred destinations to under `ship to country`.
- `Delivery Plant` had a median distribution across the dataset as well, as well as `Sales quantity` being average. 
- The majority of the `External Agents` were not assigned.
- `Price TC /Kg` of the different medical solutions were left skewed, which meant there were more higher selling drugs than cheaper drugs for sale. 
- `External commissions` were right skewed, as most of the commisions were on the lower end of the spectrum. `March` was the most frequent `month`.

These are all the noticed insights generated from the dataset through EDA (Exploratory Data Analysis).
